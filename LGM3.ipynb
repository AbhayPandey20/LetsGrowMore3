```python
# Download the image using requests
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))
```

```python
# Convert the PIL image to a numpy array
image_np = np.array(image)
```

```python
# Convert the image to RGB format (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB)
```

```python
# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
```

```python
# Invert the grayscale image
image_inverted = cv2.bitwise_not(image_gray)
```

```python
# Apply Gaussian blur to the inverted grayscale image
image_blurred = cv2.GaussianBlur(image_inverted, (21, 21), sigmaX=0, sigmaY=0)
```

```python
# Invert the blurred image
image_blurred_inverted = cv2.bitwise_not(image_blurred)
```

```python
# Create the pencil sketch by dividing the grayscale image by the inverted blurred image
sketch = cv2.divide(image_gray, image_blurred_inverted, scale=256.0)
```

```python
# Display the original image and the pencil sketch side by side
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('on')

plt.subplot(1, 2, 2)
plt.imshow(sketch, cmap='gray')
plt.title("Pencil Sketch")
plt.axis('on')

plt.show()
```
